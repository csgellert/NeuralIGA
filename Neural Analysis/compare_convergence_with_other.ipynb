## SUBDIV = 4, Trf

```python
divs = [30, 50, 80,100,120,150]
len_u = [756,1983,4902,7584,10815,16759]
all_elems = [(i+2)**2 for i in divs]

L2 = [1.884424e-02,9.844389e-03,2.191667e-03,6.896887e-04,4.001869e-04,1.492132e-04]
H1_semi = [2.434440e-01,2.701380e-02,6.078845e-02,3.056402e-02,1.566976e-02,8.072063e-03]
H1 = [2.441722e-01,1.775834e-01,6.082795e-02,3.057180e-02,1.567486e-02,8.073442e-03]
MAE = [9.398021e-04,3.032011e-04,6.766899e-05,2.397336e-05,1.992592e-05,6.494758e-06]
L_inf = [2.462320e-02,1.723439e-02,6.662728e-03,3.080421e-03,1.758471e-03,6.735824e-04]
```

```python
plt.loglog(len_u, L2, marker='o', label='L2')
plt.loglog([1e-1,1e5], [1e0,1e-12], marker='o')
plt.show()
```

```python
#fit lines and calculate slopes for L2
coeffs = np.polyfit(np.log(len_u[1:]), np.log(L2[1:]), 1)
slope = coeffs[0]
print(f"L2 slope: {slope}")
```

```python
plt.loglog(len_u, H1, marker='o', label='H1 Error')
plt.loglog([1e0,1e5], [1e0,1e-10], marker='o')
plt.show()
```

```python
#fit lines and calculate slopes for L2
coeffs = np.polyfit(np.log(len_u[1:]), np.log(H1[1:]), 1)
slope = coeffs[0]
print(f"H1 slope: {slope}")
```
