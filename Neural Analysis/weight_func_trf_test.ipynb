# Case 3 - Claasic Weighted splines

```python
x = np.array([0.01, 0.04, 0.08, 0.12, 0.16, 0.2, 0.24, 0.28, 0.32, 1.0])
y = np.array([1.89e-4, 8.01e-5, 2.76e-5, 1.42e-5, 8.72e-6, 6.97e-6,
              5.85e-6, 5.28e-6, 4.91e-6, 3.86e-6])
import matplotlib.pyplot as plt
plt.plot(x, y, marker='o')
plt.hlines(y=4.096645e-06, xmin=0.0, xmax=1.0, colors='r', linestyles='dashed', label='True SDF')
#plt.xscale('log')
plt.yscale('log')
plt.xlabel('p=2,DIV=50,gamma=3,CASE=3')
plt.show()
```
