```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.yscale('log')
plt.xscale('log')
plt.plot(esize[mask],MSE3[mask],'*-')
plt.plot(esize[mask],MSE2[mask],'*-')
plt.plot(esize[mask],MSE1[mask],'*-')
plt.show()
```
