```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#code from 2024.10.30.
# L-shape euclidian distance
test_values_euk = np.array([20,30,40,50,80])
mask1_euk = np.array([True,True,True,True,True],dtype=np.bool8)
mask_all_euk = np.ones_like(mask1_euk)
mask_euk = mask_all_euk
esize_euk = np.array([1/(nd+1) for nd in test_values_euk],dtype=np.float32)
MSE1_euk = np.array([None,2.9741983e-05,1.39898575e-05,6.664014e-06,1.7440939e-06])
MSE2_euk = np.array([2.4280105e-05,None,1.5083928e-06,None,6.6933274e-07])
MSE3_euk = np.array([None,3.2439448e-06,1.333366e-06,7.662685e-07,6.549402e-07])
etypes_euk = [{'outer': 100, 'inner': 261, 'boundary': 80},{'outer': 225, 'inner': 616, 'boundary': 120}, {'outer': 400, 'inner': 1121, 'boundary': 160},{'outer': 625, 'inner': 1776, 'boundary': 200},{'outer': 1600, 'inner': 4641, 'boundary': 320}]
inner_euk = np.array([case["inner"] for case in etypes_euk])
outer_euk = np.array([case["outer"] for case in etypes_euk])
boundary_euk = np.array([case["boundary"] for case in etypes_euk])

case_euk = {'name':'euk','MSE1':MSE1_euk,'MSE2':MSE2_euk,'MSE3':MSE3_euk,'mask':mask_euk,'test_values':test_values_euk,'esize':esize_euk,'inner':inner_euk,'outer':outer_euk,'boundary':boundary_euk}
```

```python
#code from 2024.10.30.
# L-shape kvadratic
test_values_k2 = np.array([20,30,40,50,80])
mask1_k2 = np.array([True,True,True,True,True],dtype=np.bool8)
mask_all_k2 = np.ones_like(mask1_k2)
mask_k2 = mask_all_k2
esize_k2 = np.array([1/(nd+1) for nd in test_values_k2],dtype=np.float32)
MSE1_k2 = np.array([])
MSE2_k2 = np.array([])
MSE3_k2 = np.array([None,None,None,None,None])
etypes_k2 = [{'outer': 100, 'inner': 261, 'boundary': 80},{'outer': 225, 'inner': 616, 'boundary': 120}, {'outer': 400, 'inner': 1121, 'boundary': 160},{'outer': 625, 'inner': 1776, 'boundary': 200},{'outer': 1600, 'inner': 4641, 'boundary': 320}]
inner_k2 = np.array([case["inner"] for case in etypes_k2])
outer_k2 = np.array([case["outer"] for case in etypes_k2])
boundary_k2 = np.array([case["boundary"] for case in etypes_k2])

case_k2 = {'name':'k2','MSE1':MSE1_k2,'MSE2':MSE2_k2,'MSE3':MSE3_k2,'mask':mask_k2,'test_values':test_values_k2,'esize':esize_k2,'inner':inner_k2,'outer':outer_k2,'boundary':boundary_k2}
```

```python
cases = [case_euk]
ccase = case_euk
```

```python
labels = []
for c in cases:
    plt.yscale('log')
    plt.xscale('log')
    labels.append(c['name'])
    plt.plot(c['esize'][c['mask']],c['MSE3'][c['mask']],'*r-')
    plt.plot(c['esize'][c['mask']],c['MSE2'][c['mask']],'*-')
    plt.plot(c['esize'][c['mask']],c['MSE1'][c['mask']],'*b-')
plt.legend(labels)
plt.show()
    
```
