```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

#code from 2024.10.27.
test_values = np.array([2,6,14,20,30,40,50,60])
mask = np.array([False,False,True,True,True,True,True,True],dtype=np.bool8)
esize = np.array([1/(nd+1) for nd in test_values],dtype=np.float32)
MSE1 = np.array([0.00064582,2.3073011e-05,1.4173708e-06,4.208904e-07,2.0358682e-07,1.7508637e-07,1.125297e-07,7.7746684e-08])
MSE2 = np.array([8.767728e-06,7.923716e-07,2.5704358e-07,2.005082e-07,1.5010642e-07,1.3746286e-07,9.76022e-08,5.8070047e-08])
MSE3 = np.array([3.3133745e-06,5.3852193e-07,2.5285578e-07,1.9312496e-07,1.5456618e-07,1.3048265e-07,9.845559e-08,4.830123e-08])
etypes = [{'outer': 0, 'inner': 1, 'boundary': 8}, {'outer': 4, 'inner': 21, 'boundary': 24}, {'outer': 24, 'inner': 145, 'boundary': 56}, {'outer': 68, 'inner': 293, 'boundary': 80},{'outer': 168, 'inner': 673, 'boundary': 120}, {'outer': 304, 'inner': 1217, 'boundary': 160}, {'outer': 484, 'inner': 1917, 'boundary': 200}, {'outer': 724, 'inner': 2757, 'boundary': 240}]
inner = np.array([case["inner"] for case in etypes])
outer = np.array([case["outer"] for case in etypes])
boundary = np.array([case["boundary"] for case in etypes])
```

```python
plt.yscale('log')
plt.xscale('log')
plt.plot(esize[mask],MSE3[mask],'*-')
plt.plot(esize[mask],MSE2[mask],'*-')
plt.plot(esize[mask],MSE1[mask],'*-')
plt.show()
```

```python
plt.yscale('log')
#plt.xscale('log')
plt.plot(test_values[mask]**2,MSE3[mask],'*-')
plt.plot(test_values[mask]**2,MSE2[mask],'*-')
plt.plot(test_values[mask]**2,MSE1[mask],'*-')
plt.show()
```

```python
plt.plot(esize[mask],((inner+boundary)/(inner+boundary+outer))[mask],'*-')
plt.plot(esize[mask],((inner)/(inner+boundary+outer))[mask],'*-')
plt.plot(esize[mask],((boundary)/(inner+boundary+outer))[mask],'*-')
plt.legend(['ossz','in','bnd'])
plt.show()
```

```python
plt.yscale('log')
plt.plot(esize[mask],(outer[mask]),'*-')
plt.plot(esize[mask],(inner[mask]),'*-')
plt.plot(esize[mask],(boundary[mask]),'*-')
plt.legend(['out','in','bnd'])
plt.show()
```
